### Get index name

```python
from elasticsearch import Elasticsearch
es = Elasticsearch(hosts= "http://192.168.120.185:9200")
es.ping()
```

### Query string DSL

## 1) 

```python
query = {
            "query_string": {
                "default_field": "title",
                "query": "jav?"
            }    
        }

data_search = es.search(index="books", query=query)
data_search = data_search['hits']['hits']
print("The number of",len(data_search))

for i in data_search:
    print("title :", i['_source']['title'])
```

## 2) 

```python
query = {
    "match_phrase":{
        "title": "effective"
    }
}

result = es.search(index="books", query=query)['hits']['hits']
for i in result:
    print(i['_source']['title'])
```

## 3) 

```python
query = {
    "match_phrase_prefix":{
        "title": "ja"
    }
}

result = es.search(index="books", query=query)['hits']['hits']
for i in result:
    print(i['_source']['title'])
```

### Create Index

```python
result = es.indices.create(index="test", mappings={
        "properties":
        {"name":{"type":"text"},
         "family":{"type":"text"}
        }
   })
    
print(result['acknowledged'])
```

### Add document

```python
res = es.index(index="test", id=123, document={"name":"sanaz" ,"family":"khalili"})
res
```

## bulk

```python
from elasticsearch import helpers

action = [{"_index": "books", "title": "my_book", "author": "sanaz", "edition":1},
          {"_index": "books", "title": "my_book2", "author": "sanaz2", "edition":2}]
helpers.bulk(es, action )
```
